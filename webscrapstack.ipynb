```python
import requests
from bs4 import BeautifulSoup
```

```python
url=requests.get('https://stackoverflow.com/questions/415511/how-do-i-get-the-current-time-in-python',headers={"User-Agent": "Mozilla/5.0"})
print(url.text)
```

```python
soup= BeautifulSoup(url.text,"html.parser")
print(soup.prettify())
```

```python
question=soup.find('a',class_="question-hyperlink")
print("Question:",question.text)
```

```python
answers=soup.find_all('div',class_="s-prose js-post-body")
anum=1
for i in answers[1:]:
    print(f"{anum} Answer:{i.text}")
    anum+=1
```
